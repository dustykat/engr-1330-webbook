```python
import pandas as pd

numRolls = []
probabilities = []
for i in range(50):
    numRolls.append(i)
    probabilities.append(1-(5/6)**i)

rolls = {
    "NumRolls": numRolls,
    "Prob at least 6": probabilities
}
```

```python
df = pd.DataFrame(rolls)
df.plot.scatter(x="NumRolls", y="Prob at least 6")
```
